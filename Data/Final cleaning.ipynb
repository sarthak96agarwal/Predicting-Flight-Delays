```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('FinalData.csv',index_col=0)
df.head()
```

```python
df=df.drop(['DATE','TAIL_NUM'],axis=1)
```

```python
df.columns
```

```python
df['DEP_TIME']=df['DEP_TIME']//100
```

```python
df=df.drop('DEP_DELAY',axis=1)
```

```python
df=df.drop(['TMAX','TMIN'],axis=1)
df
```

```python
carrier=pd.get_dummies(df['OP_UNIQUE_CARRIER'])
```

```python
merged=df.merge(carrier,on=df.index)
```

```python
merged=merged.drop(['OP_UNIQUE_CARRIER','key_0'],axis=1)
```

```python
merged.columns
```

```python
merged=merged.drop('WSF2',axis=1)
```

```python
merged=merged.rename(columns={'DAY_OF_MONTH':'DAY','DELAY':'Y','AWND':'WIND_SPEED','PRCP':'RAIN','SNWD':'SNOW_DEPTH','WT01':'FOG','WT02':'HEAVY_FOG','WT03':'THUNDER','WT04':'HAIL','WT06':'GLAZE','WT08':'HAZE','WT09':'BLOWING_SNOW','holiday':'HOLIDAY'})
```

```python
merged.columns
```

```python
merged.to_csv('Final.csv')
```

```python
from sklearn.model_selection import train_test_split
import random
random.seed(0)
X_train,X_test,y_train,y_test=train_test_split(merged.drop('Y',axis=1),merged['Y'],test_size=0.25)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```
