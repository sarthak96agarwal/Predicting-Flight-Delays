```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('2016Aug-2017Jul.csv')
df2=pd.read_csv('2017Aug-2018Jul.csv')
```

```python
df=pd.concat([df1,df2])
```

```python
df=df.drop('Unnamed: 0',axis=1)
df=df.drop('Unnamed: 9',axis=1)
df=df.drop(['ORIGIN_AIRPORT_ID','ORIGIN_CITY_MARKET_ID'],axis=1)
df['MONTH'] = df['MONTH'].apply(lambda x: '{0:0>2}'.format(x))
df['DAY_OF_MONTH'] = df['DAY_OF_MONTH'].apply(lambda x: '{0:0>2}'.format(x))
df["DATE"] = df['YEAR'].map(str)+'-'+df["MONTH"].map(str) +'-'+ df['DAY_OF_MONTH'].map(str)
df=df.reset_index()
df
```

```python
weather=pd.read_csv("Weather_Data.csv")
```

```python
merged_df=df.merge(weather,on='DATE')
merged_df
```

```python
merged_df.to_csv('merged.csv')
```
